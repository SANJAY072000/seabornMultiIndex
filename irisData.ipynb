```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
sns.distplot(df['petal_length'])
```

```python
sns.pairplot(df)
```

```python
sns.jointplot('petal_length','petal_width',df)
```

```python
df[(df['petal_length']>2.5)&(df['petal_length']<3.5)]
```

```python
sns.factorplot('petal_length','petal_width',data=df,col='species')
```

```python
sns.swarmplot('species','petal_length',data=df)
```

```python
plt.plot(df['petal_length'],df['petal_width'],label='Realation plot')
plt.legend()
```
